##### **Step 3: Implement the Recommendation Function**
Now, implement the function `recommend_movies_for_user(user_id, num=5)` to recommend movies for a given user.

##### **Function Inputs:**
- `user_id`: The target user for whom we need recommendations.
- `num`: The number of movies to recommend (default is 5).

##### **Function Steps:**
1. Find **similar users**:
   - Retrieve the similarity scores for the given `user_id`.
   - Sort them in **descending** order (highest similarity first).
   - Exclude the user themselves.
   
2. Get the **movie ratings** from these similar users.

3. Compute the **average rating** for each movie based on these users' preferences.

4. Sort the movies in **descending order** based on the computed average ratings.

5. Retrieve the **top `num` recommended movies**.

6. Map **movie IDs** to their **titles** using the `movies` DataFrame.

7. Return the results as a **Pandas DataFrame** with rankings.

````python
# Imports
import pandas as pd  # For handling data
import numpy as np   # For numerical computations
from sklearn.metrics.pairwise import cosine_similarity  # For computing user similarity

# Recommendation Function
#      Inputs -> user_id (int: id of user to recommend to),
#                num (int: number of recommendations),
#                threshold (int: optional number of max number of similar users to compare to)
#      Outputs -> results_df (Pandas DataFrame)
def recommend_movies_for_user(user_id, num=5, threshold=5):

  # Check to see if user_id is in user_movie_matrix
  if user_id not in user_movie_matrix.index:
    return "User not found"
  # Check to see is number of recommendations wanted is larger than the number of recommendable movies
  if num > len(user_movie_matrix):
    num = len(user_movie_matrix)
  # Check to see if valid number of recommendations was asked
  if num <= 0:
    return "Invalid number of recommendations"

  # Replace NaN Values with 0 + Compute Cosine Similarity
  user_similarity = cosine_similarity(user_movie_matrix.fillna(0))
  user_sim_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

  # Find Similar Users + Sort in Descending + Remove Row w/Current User
  user_sims = user_sim_df[user_id].drop(user_id).sort_values(ascending=False)

  # Reduce Similar Users Down to The Top (Threshold) Similar Users
  user_sims = user_sims.head(threshold)

  # Get Movie Ratings From Similar Users
  movie_ratings = user_movie_matrix.loc[user_sims.index]

  # Compute Average Rating
  avg_ratings = movie_ratings.mean().dropna()

  # Sort Movies
  sorted_ratings = avg_ratings.sort_values(ascending=False)

  # Get movies already rated by user
  rated_movies = user_movie_matrix.loc[user_id].dropna().index

  # Filter out already rated movies
  sorted_avg_ratings = sorted_ratings[~sorted_ratings.index.isin(rated_movies)]

  # Retrieve Top Movies + Convert To DataFrame
  top_movies = pd.DataFrame(sorted_avg_ratings.head(num))
  top_movies['movie_id'] = top_movies.index

  # Map Movie IDs
  top_movie_ids = top_movies['movie_id'] - 1

  movie_names = movies_cleaned.loc[top_movie_ids]['title']

  # Create Results DataFrame
  results_df = pd.DataFrame({
      'Ranking': range(1, num+1),
      'Movie Name': movie_names
  })
  results_df.set_index('Ranking', inplace=True)

  return results_df

# Test
recommend_movies_for_user(10, num = 5)
````

````python
# Imports
import pandas as pd  # For handling data
import numpy as np   # For numerical computations
from sklearn.metrics.pairwise import cosine_similarity  # For computing user similarity

# Recommendation Function
#      Inputs -> movie_name (str: name of movie to recommend for),
#                num (int: number of recommendations)
#      Outputs -> results_df (Pandas DataFrame)
def recommend_movies(movie_name, num=5):

  # Replace NaN Values with 0 + Compute Cosine Similarity
  item_similarity = cosine_similarity(user_movie_matrix.T.fillna(0))
  item_sim_df = pd.DataFrame(item_similarity, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)

  # Find movie_id
  movie = movies_cleaned[movies_cleaned['title'] == movie_name]

  # Edge Case Checker
  if movie.empty:
    return "Movie not found"
  # Check if there multiple movies were found with the same name
  if len(movie) > 1:
    return "Multiple movies with same name"
   # Check to see is number of recommendations wanted is larger than the number of recommendable movies
  if num > len(item_sim_df):
    num = len(item_sim_df)
  # Check to see if a valid number of recommendations was asked
  if num <= 0:
    return "Invalid number of recommendations"

  movie_id = movie.iloc[0]['movie_id']

  # Extract similarity scores for movie
  movie_sims = item_sim_df[movie_id]

  # Sort by descending order
  movie_sims = movie_sims.drop(movie_id).sort_values(ascending=False)

  # Retrieve Top Movies + Convert To DataFrame
  top_movies = pd.DataFrame(movie_sims.head(num))
  top_movies['movie_id'] = top_movies.index

  # Map Movie IDs (reduces all movie_ids by 1 since the actual ids and the index are off by one)
  top_movie_ids = top_movies['movie_id'] - 1

  movie_names = movies_cleaned.loc[top_movie_ids]['title']

  # Create Results DataFrame
  results_df = pd.DataFrame({
      'Ranking': range(1, num+1),
      'Movie Name': movie_names
  })
  results_df.set_index('Ranking', inplace=True)

  return results_df

# Test
recommend_movies("Jurassic Park (1993)", num=5)
````

## **Part 3: Graph-Based Recommender (Pixie-Inspired Algorithm)**

### **Adjacency List**

#### **Objective**
In this task, you will preprocess the Movie dataset and construct a **graph representation** where:
- **Users** are connected to the movies they have rated.
- **Movies** are connected to users who have rated them.
  
This graph structure will help in exploring **user-movie relationships** for recommendations.

#### **Step 1: Merge Ratings with Movie Titles**
Since we have **movie IDs** in the ratings dataset but need human-readable movie titles, we will:
1. Merge the `ratings` DataFrame with the `movies` DataFrame using the `'movie_id'` column.
2. This allows each rating to be associated with a **movie title**.

#### **Hint:**
Use the following Pandas operation to merge:



#### **Step 2: Aggregate Ratings**
Since multiple users may rate the same movie multiple times, we:
1. Group the dataset by `['user_id', 'movie_id', 'title']`.
2. Compute the **mean rating** for each movie by each user.
3. Reset the index to ensure we maintain a clean DataFrame structure.

#### **Hint:**  
Use `groupby()` and `mean()` as follows:
```python

```

#### **Step 3: Normalize Ratings**
Since different users have different rating biases, we normalize ratings by:
1. **Computing each user's mean rating**.
2. **Subtracting the mean rating** from each individual rating.

#### **Instructions:**
- Use `groupby('user_id')` to group ratings by users.
- Apply `transform(lambda x: x - x.mean())` to adjust ratings.

#### **Hint:**  
Normalize ratings using:
```python
ratings['rating'] = ratings.groupby('user_id')['rating'].transform(lambda x: x - x.mean())
```
This ensures each user’s ratings are centered around zero, making similarity calculations fairer.

#### **Step 4: Construct the Graph Representation**
We represent the user-movie interactions as an **undirected graph** using an **adjacency list**:
- Each **user** is a node connected to movies they rated.
- Each **movie** is a node connected to users who rated it.

#### **Graph Construction Steps:**
1. Initialize an empty dictionary `graph = {}`.
2. Iterate through the **ratings dataset**.
3. For each `user_id` and `movie_id` pair:
   - Add the movie to the user’s set of connections.
   - Add the user to the movie’s set of connections.

#### **Hint:**  
The following code builds the graph:

```python
graph = {}
for _, row in ratings.iterrows():
    user, movie = row['user_id'], row['movie_id']
    if user not in graph:
        graph[user] = set()
    if movie not in graph:
        graph[movie] = set()
    graph[user].add(movie)
    graph[movie].add(user)
```

This results in a **bipartite graph**, where:
- **Users** are connected to multiple movies.
- **Movies** are connected to multiple users.

#### **Step 5: Understanding the Graph**
- **Nodes** in the graph represent **users and movies**.
- **Edges** exist between a user and a movie **if the user has rated the movie**.
- This structure allows us to find **users with similar movie tastes** and **movies frequently watched together**.

#### **Exploring the Graph**
- **Find a user’s rated movies:**  
  ```python
  user_id = 1
  print(graph[user_id])  # Movies rated by user 1
  ```

- **Find users who rated a movie:**  
  ```python
  movie_id = 50
  print(graph[movie_id])  # Users who rated movie 50
  ```

````python
# Merge Ratings & Movies
ratings_movies = ratings_cleaned.merge(movies_cleaned, on='movie_id')

# Aggregate Ratings
ratings_movies = ratings_movies.groupby(['user_id', 'movie_id', 'title'])['rating'].mean().reset_index()

# Normalize Ratings
ratings_movies['rating'] = ratings_movies.groupby('user_id')['rating'].transform(lambda x: x - x.mean())

# Construct Graph
graph = {}
for _, row in ratings_movies.iterrows():
    user, movie = row['user_id'], row['movie_id']

    user_key = f'user_{user}'
    movie_key = f'movie_{movie}'

    if user_key not in graph:
        graph[user_key] = set()
    if movie_key not in graph:
        graph[movie_key] = set()
    graph[user_key].add(movie_key)
    graph[movie_key].add(user_key)
````

````python
# Imports
import random
import pandas as pd

# Global Constants For Testing/Debugging
BETA = 1
DEBUG_MOVIES = ['Jurassic Park (1993)','Night Falls on Manhattan (1997)','Star Trek VI: The Undiscovered Country (1991)','Commandments (1997)','E.T. the Extra-Terrestrial (1982)']
DEBUG_WALK = 1500000
DEBUG_NUM = 10
DEBUG_VISIT = 100

def weighted_pixie_recommend(movie_name, walk_length=15, num=5):
  # Retrieve movie for given movie_name
  movie = movies_cleaned[movies_cleaned['title'] == movie_name]

  # Check to see if movie exists for given movie_name
  if movie.empty:
    return "Movie not found"

  # Check to see if walk length is valid number
  if walk_length <= 0 or walk_length < num:
    return "Invalid walk length"

  # Check to see if num is a valid number
  if num <= 0:
    return "Invalid number of recommendations"

  # Check to see if BETA it too high or low
  if BETA < 0 or BETA > 1:
    return "Invalid beta value"

  # Retrieve id of given movie
  seed_id = movie.iloc[0]['movie_id']

  # Check to see if movie_id exists in graph
  if f'movie_{seed_id}' not in graph:
    return "Movie not found"

  # Generate start node
  movie_node = graph[f'movie_{seed_id}']
  visited_movies = {}

  # Start looping
  for _ in range(walk_length):
    # Walk to user
    next_user = random.choice(list(movie_node))
    user_node = graph[next_user]

    # Walk to movie
    next_movie = random.choice(list(user_node))
    movie_node = graph[next_movie]

    # Retrieve movie id
    movie_id = int(next_movie.split('_')[1])

    # Continue if random walk has naturally restarted
    if movie_id == seed_id:
      continue

    # Keep track of movie visitation numbers
    if movie_id in visited_movies:
      visited_movies[movie_id] += 1

      # Stop random walk if a movie has reached 20 visitations
      if visited_movies[movie_id] >= DEBUG_VISIT:
        break

    else:
      visited_movies[movie_id] = 1

    # Simulate restarts
    if random.random() < (1 - BETA):
      movie_node = graph[f'movie_{seed_id}']

  # Sort visited movies
  sorted_movies = sorted(visited_movies.items(), key=lambda x: x[1], reverse=True)[:num]

  # Check to see if enough movies were visited
  if len(sorted_movies) < num:
    return "Not enough movies were visited"

  # Convert sorted_movies into DataFrame
  sorted_movies = pd.DataFrame(sorted_movies, columns=['movie_id', 'visits'])

  # Map Movie IDs
  top_movie_ids = sorted_movies['movie_id']

  movie_names = movies_cleaned.loc[top_movie_ids]['title']

  # Create Results DataFrame
  results_df = pd.DataFrame({
      'Ranking': range(1, num+1),
      'Movie Name': movie_names
  })
  results_df.set_index('Ranking', inplace=True)

  return results_df

weighted_pixie_recommend(DEBUG_MOVIES[0], walk_length=DEBUG_WALK, num=DEBUG_NUM)
````
